### EDA

# Histograms for remaining column

```python
df[df.columns[:9]].hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
# I improve labels and grid lines and distinct colors for each and organized in drid layout
# Set figure size
fig, axes = plt.subplots(3, 3, figsize=(15, 10))  # 3x3 grid for 9 features
axes = axes.flatten()  # Flatten for easy indexing

# Define color palette for better distinction
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'cyan', 'brown', 'gray']

# Loop through the first 9 columns to create histograms
for i, col in enumerate(df.columns[:9]):
    df[col].hist(ax=axes[i], bins=25, color=colors[i], edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Frequency', fontsize=10)
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)  # Adding grid for better readability

# Adjust layout to prevent overlap
plt.tight_layout()
plt.suptitle('Feature Distributions', fontsize=16, fontweight='bold', y=1.02)  # Title for entire plot
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['generated_power_kw'], color='orange')
plt.title('Scatter Plot of Generated Power (kW)')
plt.xlabel('Index')
plt.ylabel('Generated Power (kW)')
plt.grid(True)
plt.show()
```

```python
# Filter only numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns.tolist()

# Remove target column if it exists
if 'generated_power_kw' in numeric_columns:
    numeric_columns.remove('generated_power_kw')

# Check if there are numeric columns left
if not numeric_columns:
    print("No numeric columns available for plotting.")
else:
    # Dynamically adjust subplot grid
    rows = (len(numeric_columns) // 3) + 1  # Calculate required rows

    plt.figure(figsize=(15, 5 * rows))  # Adjust figure size based on rows

    for i, column in enumerate(numeric_columns, start=1):  # Start index from 1
        plt.subplot(rows, 3, i)
        plt.scatter(df[column], df['generated_power_kw'], alpha=0.5, c='blue')
        plt.title(f'{column} vs Generated Power (kW)')
        plt.xlabel(column)
        plt.ylabel('Generated Power (kW)')

    plt.tight_layout()
    plt.show()
```

```python
# Filters only numeric columns using df.select_dtypes(include=['number'])
# Better color representation using color='skyblue'
# Prevents clutter by using plt.tight_layout()
# Ensures proper spacing by using rotation=45

# Set style for better readability
sns.set_style('whitegrid')

# Create a figure for all boxplots
plt.figure(figsize=(15, 30))

# Loop through numerical columns and plot boxplots
for i, column in enumerate(df.select_dtypes(include=['number']).columns, 1):
    plt.subplot(7, 3, i)
    sns.boxplot(x=df[column], color='skyblue')
    plt.title(f'Boxplot of {column}')
    plt.xticks(rotation=45)

# Adjust layout and display plots
plt.tight_layout()
plt.show()
```
